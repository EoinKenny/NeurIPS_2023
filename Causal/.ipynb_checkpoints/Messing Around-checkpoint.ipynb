```python
import numpy as np
import os
import pandas as pd
import seaborn as sns

from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
datasets = ['compas', 'adult']
methods = ['us', 'karimi', 'dominguez']
seeds = [0,1,2,3,4]
```

```python
def avg_col(temp):   
    temp_row = pd.DataFrame()

    for thing in temp.columns:
        try:
            temp_row[thing] = [temp[thing].mean()]
        except:
            temp_row[thing] = [temp[thing][0]]

    return temp_row
```

```python
data1 = pd.DataFrame()

for d in datasets:
    for m in methods:
        for s in seeds:
            temp = pd.read_csv('results/' + d + m + str(s) + '.csv')
            temp = avg_col(temp)
            data1 = pd.concat([data1, temp])
        
data2 = pd.DataFrame()

for d in datasets:
    for m in methods:
        for s in seeds:
            temp = pd.read_csv('results/' + d + m + str(s) +'.csv')
            temp = avg_col(temp)
            data2 = pd.concat([data2, temp])
```

```python
data1['GainType'] = 'Action Gain'
del data1['Gain']
data1 = data1.rename(columns={'CostAction': 'Gain'})
data2['GainType'] = 'Causal Gain'
del data2['CostAction']

data = pd.concat([data1, data2])
```

```python
data = data.rename(columns={"RobustnessAdv": "Robustness"})
data = data.replace('compas', 'Compas',)
data = data.replace('adult', 'Adult', )

data = data.replace('us', 'S-GEN', )
data = data.replace('karimi', 'Karimi et al. (2021)*', )
data = data.replace('dominguez', 'Dominguez et al. (2022)*', )
```

## Standard Error for Paper
